```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv")
vine_df.head()
```

```python
total_votes_df = vine_df[(vine_df)["total_votes"] >= 20]
total_votes_df.head()
```

```python
helpful_votes_df = total_votes_df[total_votes_df['helpful_votes']/total_votes_df['total_votes'] >= 0.5]
```

```python
helpful_votes_df
```

```python
vine_yes_df = helpful_votes_df[(helpful_votes_df)["vine"] == 'Y']
vine_yes_df
```

```python
vine_no_df = helpful_votes_df[(helpful_votes_df)["vine"] == 'N']
vine_no_df
```

```python
vine_yes_df.count()
```

```python
helpful_votes_df.count()
```

```python
vine_yes_df.dtypes
```

```python
# total reviews for Y vine
total_vine_yes = vine_yes_df["star_rating"].count()
```

```python
# 5 star ratings for VINE Y 
vine_yes_5star_reviews = vine_yes_df.loc[(vine_yes_df['star_rating']== 5)]["star_rating"].count()
print(vine_yes_5star_reviews)
```

```python
# % of 5 * reviews for Y vine
percentage_5star_reviews_vine_y = vine_yes_5star_reviews / total_vine_yes * 100
print(percentage_5star_reviews_vine_y)
```

```python
# total reviews for N vine
total_vine_no = vine_no_df["star_rating"].count()
```

```python
# 5 star ratings for VIne N
vine_no_5star_reviews = vine_no_df.loc[(vine_no_df['star_rating']== 5)]["star_rating"].count()
print(vine_no_5star_reviews)
```

```python
# % of 5 * reviews for N vine
percentage_5star_reviews_vine_n = vine_no_5star_reviews / total_vine_no * 100
print(percentage_5star_reviews_vine_n)
```
